# Métrica - Mean Absolute Error (MAE)

```python
valores_reais = [2000, 2150, 2500, 1990, 2220]

valores_preditos = [2100, 2450, 2730, 1990, 2140]
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(valores_reais, valores_preditos)
```

```python
print(mae)
```

## Visualização Gráfica

```python
import matplotlib.pyplot as plt

def plot_scatter(valores_reais, valores_preditos):
    plt.figure(figsize=(8, 6))
    plt.scatter(valores_reais, valores_preditos, color='blue')
    plt.plot(valores_reais, valores_reais, color='red', linestyle='--')
    plt.title('Valores Reais vs. Valores Preditos')
    plt.xlabel('Valores Reais')
    plt.ylabel('Valores Preditos')
    plt.grid(True)
    plt.show()

plot_scatter(valores_reais, valores_preditos)
```
